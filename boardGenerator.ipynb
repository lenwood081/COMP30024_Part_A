```python
import csv
import random

def generate_boards():
    boards = []
    base_board = [[' ' for _ in range(8)] for _ in range(8)]
    base_board[0][4] = 'r'  # Ensure 'r' is always in the top row
    
    added_positions = [(0, 4)]  # Start with 'r' position
    fill_order = [(r, c) for r in range(8) for c in range(8) if (r, c) != (0, 4)]  
    random.shuffle(fill_order)  # Randomize the fill order

    for i in range(1, 65):  # Generate 64 different boards
        new_board = [row[:] for row in base_board]  # Copy board
        if i < len(fill_order):
            added_positions.append(fill_order[i-1])  # Add a new character

        for (r, c) in added_positions:
            new_board[r][c] = '*' if random.random() < 0.75 else 'b'  # * is 3x more likely than b
        
        new_board[0][4] = 'r'  # Ensure 'r' is always in the top row
        boards.append(new_board)

    return boards

def save_boards_to_csv(boards):
    for i, board in enumerate(boards):
        filename = f"board_{i+1}.csv"
        with open(filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(board)
        print(f"Saved: {filename}")

# Generate and save the boards
boards = generate_boards()
save_boards_to_csv(boards)

print("All CSV files have been generated!")
```
